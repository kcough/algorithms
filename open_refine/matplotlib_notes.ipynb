```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('unemployment.csv')
df.head()
```

## when you use df.plot() you're plotting your index vs your columns 
## if you have a series with a meaningful index you can do .plot()

```python
#so since you don't want to use the index
#you have to tell it which columns to use
#but this doesn't understand that all the countries are different
#so it plots one country then it goes back (note the backwards line)
#and plot the thing. it assumes that ALL OF THE DATA IS IN ORDER
#except it isn't. you have data for the usa then australia
df.plot(x='year', y='unemployment', kind='bar')
```

```python
df.groupby('country').plot(x='year',y='unemployment')
```

```python
import matplotlib.pyplot as plt
#pyplot is the thing inside of matplotlib that does everything
#it's the same as talking about matplotlib

fig, ax = plt.subplots()
#fig means figure. your canvas that you're drawing on
#talking about a graph call it an axis, or a subplot. it's all the same
#we're saying graphing library, make us a new graph:
#it'll give us a figure (our whole canvas) and one graph (ax)
#so we're storing those two things in the variables

df.groupby('country').plot(x='year',y='unemployment', ax=ax)
#we're saying, we already have an empty graph,
#use this empty graph and draw on it
```

```python
#plt, createa  figure on this page
fig = plt.figure()

#then create a new subplot on that page.
#we have to tell it 111
#111 means a 1x1 grid and give me the first part of it
#this is shorthand for 1,1,1
#you can change this! try 211
ax = fig.add_subplot(111)
ax2 = fig.add_subplot(211)

df.groupby('country').plot(x='year',y='unemployment', ax=ax)
df.groupby('country').plot(x='year',y='unemployment', ax=ax2)
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax2 = fig.add_subplot(122)
#give me the first section
#give me the second section

df.groupby('country').plot(x='year',y='unemployment', ax=ax)
df.groupby('country').plot(x='year',y='unemployment', ax=ax2)
```

```python
fig, ax = plt.subplots()

ax.plot(df.year, df.unemployment)
```

```python
fig, ax = plt.subplots()

colors = {
    'Australia':'red',
    'Germany':'blue',
    'Japan':'pink',
    'USA':'orange'
}

#for loop with groupby gives you the category name
#as well as a tiny dataframe of just that group

for country, group_df in df.groupby('country'):
    #define line colors that are connected to your colors
    line_colors = colors[country]
    group_df.plot(x='year', y='unemployment', ax=ax, label=country, color=line_color)
```
