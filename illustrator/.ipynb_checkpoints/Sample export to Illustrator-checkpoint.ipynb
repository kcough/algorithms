## Making a boring graph

```python
df.plot(x='GDP_per_capita', 
        y='life_expectancy', 
        kind='scatter', 
        xlim=(0, 70000), 
        ylim=(30,85), 
        linewidth=0, 
        figsize=(10,6))

plt.savefig('boring-graph.pdf')
```

## Fun sizes

```python
# Divide the population by a number to turn it into the circle size
sizes = df.apply(lambda row: row['Population'] / 500000, axis=1)

df.plot(x='GDP_per_capita', 
        y='life_expectancy', 
        kind='scatter', 
        xlim=(0, 70000), 
        ylim=(30,85), 
        linewidth=0, 
        alpha=0.5,
        s=sizes,
        figsize=(10,6))
```

## Grouped and colored

This gets crazy, we `groupby` and then loop. Note that we use `numpy.random.rand(3,)` to generate a random color for each plot.

```python
import numpy

fig, ax = plt.subplots(figsize=(10,6))

# Group and loop through
# 'category' is the name of the continent
# selection is the dataframe for that country
for category, selection in df.groupby('Continent'):
    # Divide the population by a number to turn it into the circle size
    sizes = selection.apply(lambda row: row['Population'] / 500000, axis=1)
    # Pick a random color
    random_color = numpy.random.rand(3,)
    
    selection.plot(x='GDP_per_capita',
                   y='life_expectancy',
                   kind='scatter',
                   s=sizes,
                   c=random_color,
                   label=category,
                   alpha=0.5,
                   ylim=[30, 85],
                   xlim=[0, 70000],
                   ax=ax)
    
ax.set_ylabel("Life expectancy")
ax.set_xlabel("Per capita GDP")
ax.legend(loc='lower right')
```
